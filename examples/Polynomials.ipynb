Learning Polynomials
==================

We will try to learn some polynomial with increasing degree, we note that:

- polynomials can be accurately learned with small word sets (`learn_resolution=2`)
- the number of states of the resulting WFA is the degree + 1

```python
%matplotlib inline

from pyautospec import FunctionWfa
```

```python
# learn 3x + 2 in the [-10,+10] interval
p1_a = FunctionWfa().fit(lambda x: 3*x + 1, x0=-10.0, x1=10.0, learn_resolution=2)

p1_a
```

```python
p1_a.model.diagram()
```

```python
p1_a.comparison_chart()
```

```python
# learn 3x^2 + 2x + 2 in the [-10,+10] interval
p2_a = FunctionWfa().fit(lambda x: 3*x**2 + 2*x + 2, x0=-10.0, x1=10.0, learn_resolution=2)

p2_a
```

```python
p2_a.model.diagram()
```

```python
p2_a.comparison_chart()
```

```python
# learn 5x^3 - 3x^2 + 2x + 2 in the [-10,+10] interval
p3_a = FunctionWfa().fit(lambda x: 5*x**3 - 3*x**2 + 2*x + 2, x0=-10.0, x1=10.0, learn_resolution=2)

p3_a
```

```python
p3_a.model.diagram()
```

```python
p3_a.comparison_chart()
```
