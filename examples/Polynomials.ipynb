Learning Polynomials
==================

```python
import numpy as np
import matplotlib.pyplot as plt

from pyautospec import FunctionWfa
```

```python
# learn 3x + 2 in the [-10,+10] interval
p1_a = FunctionWfa(lambda x: 3*x + 1, x0=-10.0, x1=10.0)

p1_a.learn(resolution=2)

p1_a.model.diagram()
```

```python
x, f, r = p1_a.reconstruct(resolution=5)

plt.plot(x, f, label="original")
plt.plot(x, r, label="reconstructed")
plt.legend()
```

```python
# reconstruction error
np.average(np.square(np.array(r) - np.array(f)))
```

```python
# learn 3x^2 + 2x + 2 in the [-10,+10] interval
p2_a = FunctionWfa(lambda x: 3*x**2 + 2*x + 2, x0=-10.0, x1=10.0)

p2_a.learn(resolution=2)

p2_a.model.diagram()
```

```python
x, f, r = p2_a.reconstruct(resolution=5)

plt.plot(x, f, label="original")
plt.plot(x, r, label="reconstructed")
plt.legend()
```

```python
# reconstruction error
np.average(np.square(np.array(r) - np.array(f)))
```

```python
# learn 5x^3 - 3x^2 + 2x + 2 in the [-10,+10] interval
p3_a = FunctionWfa(lambda x: 5*x**3 - 3*x**2 + 2*x + 2, x0=-10.0, x1=10.0)

p3_a.learn(resolution=3)

p3_a.model.diagram()
```
