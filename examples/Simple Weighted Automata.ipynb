Simple Weighted Automata
======================

```python
from pyautospec import Wfa
```

An automata that counts letters
---------------------------------------------

```python
alphabet = ["a", "b"]

# create an automaton with two states
counter = Wfa(alphabet, 2)

# 0 is the start state
counter.set_starting_state(0)

# transitions
counter.add_transition(0, "a", 1.0, 0)
counter.add_transition(0, "b", 1.0, 0)

counter.add_transition(0, "b", 1.0, 1)

counter.add_transition(1, "a", 1.0, 1)
counter.add_transition(1, "b", 1.0, 1)

# 1 is the accepting state
counter.set_accepting_state(1)

# the diagram
counter.diagram()
```

```python
counter.evaluate("a")
```

```python
counter.evaluate("aa")
```

```python
counter.evaluate("aaa")
```

```python
counter.evaluate("aaba")
```

```python
counter.evaluate("aabaaaaaaab")
```
