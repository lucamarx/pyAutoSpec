Automata Learning
================

```python
from pyautospec import SpectralLearning
```

```python
alphabet = ["a", "b"]

# instantiate a learner, here 3 is the length of the prefixes/suffixes used during learning
learn = SpectralLearning(alphabet, 3)
```

```python
# a function that counts 'b's
def count_b(s):
    return len([x for x in s if x == "b"])
```

```python
# learn the count_a function
counter = learn.learn(count_b)

counter.diagram()
```

```python
"{:.0f}".format(counter("a"))
```

```python
"{:.0f}".format(counter("aa"))
```

```python
"{:.0f}".format(counter("aabaaa"))
```
