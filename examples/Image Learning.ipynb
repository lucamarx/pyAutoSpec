Image Learning
=============

```python
from skimage import io, data
from pyautospec import ImageWfa
```

```python
# get a test image
camera = data.camera()

io.imshow(camera)
```

```python
img_wfa = ImageWfa(camera, learn_resolution=4)

img_wfa
```

```python
img, err = img_wfa.reconstruct(resolution=7)

print("reconstruction error: {:.4f}".format(err))

io.imshow(img)
```
